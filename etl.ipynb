# Extract

```python
import requests
url = "http://ixa2.si.ehu.es/stswiki/images/4/48/Stsbenchmark.tar.gz"
r = requests.get(url, allow_redirects=True)
open('./data/STSbenchmark.tar.gz', 'wb').write(r.content)
```

```python
import tarfile
file = tarfile.open('./data/STSbenchmark.tar.gz')
file.extractall('./data/STSbenchmark')
```
